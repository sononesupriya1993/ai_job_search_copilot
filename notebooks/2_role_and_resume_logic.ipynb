```python
import pandas as pd
import sys

sys.path.append("../src")

from skill_extractor import extract_skills
from role_classifier import classify_role
```

```python
df = pd.read_csv("../data/sample_inputs/sample_job_descriptions.csv")
df.head()
```

```python
df["extracted_skills"] = df["job_description"].apply(extract_skills)
df["predicted_role"] = df["job_description"].apply(classify_role)

df[["job_id", "predicted_role", "extracted_skills"]]
```

```python
from resume_advisor import generate_resume_advice
```

```python
df["resume_advice_output"] = df.apply(
    lambda row: generate_resume_advice(
        predicted_role=row["predicted_role"],
        extracted_skills=row["extracted_skills"]
    ),
    axis=1
)
```

```python
df["resume_advice_output"].iloc[0]
```

```python
df["matched_skills"] = df["resume_advice_output"].apply(lambda x: x["matched_skills"])
df["missing_keywords"] = df["resume_advice_output"].apply(lambda x: x["missing_keywords"])
df["resume_advice"] = df["resume_advice_output"].apply(lambda x: x["resume_advice"])

df[["job_id", "predicted_role", "matched_skills", "missing_keywords", "resume_advice"]]
```
